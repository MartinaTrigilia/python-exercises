#**Poems on a file**

Read all the poems from the poems directory and save them in a single file called raccolta_poesie.txt, which should be located in a new directory called "raccolta". Insert a counter before each poem (e.g., POESIA 1).

```python
import os
```

```python
os.getcwd()
```

```python
for f in os.listdir():
  print(f)
```

```python
from os import listdir,mkdir,chdir

mkdir("raccolta")
PATH = os.getcwd()
NEW_PATH = os.getcwd() + '/raccolta'

new_file = NEW_PATH + '/raccolta_poesie.txt'
new_file = open(new_file,"w+")

i = 1
for f in listdir():
  if f.endswith(".txt"):
    with open(PATH+'/'+f) as file_txt:
      new_file.write("POESIA %d\n"%i)
      new_file.write(file_txt.read())
      new_file.write("\n")

      i += 1
new_file.close()
print(new_file)

```

#How long until your birthday?
Enter your birthday in month/day format (e.g., 11/06 for June 11th). Write a program that calculates the number of days until your birthday and prints the result.

```python
from datetime import datetime
```

```python
def days_count(birth_str):
  print(birth_str)
  now = datetime.now()
  birth_dt = datetime.strptime(birth_str,'%d/%m').replace(year = now.year)
  diff = birth_dt - now
  return diff.days

b = input("write your birth day in DD/MM format:")
print(f"{days_count(b)} days until your birthday")
```

```python
import math

def sigmoide(z):
  return 1/(1+math.pow(math.e,-z))

def ReLU(z):
  if z < 0:
    return 0
  else:
    return z

def tanh(z):
  res = (1-math.pow(math.e,-2*z))/(1+math.pow(math.e,-2*z))
  return res

def log_loss(y,a):
  s = 0
  for y_i,a_i in zip(y,a):
    partial_sum = (y_i * math.log(a_i)) + (1 - y_i)*math.log(1-a_i)
    s = s + partial_sum
  return (-1/len(y))*s

print(sigmoide(1))
print(ReLU(1))
print(tanh(1))
print(log_loss([1,0,1,1,0], [0.75, 0.4, 0.6, 0.8, 0.1]))
```
