## **1. La Circonferenza**
Definisci una classe per rappresentare una circonferenza che, prendendo in input il raggio, definisca dei metodi per il calcolo di diametro, circonferenza e area. Documenta i metodi utilizzando le Docstring.
Il diametro (d) di un cerchio è il doppio del raggio (\(d=2r\)), la circonferenza (C) (il perimetro) è data da \(C=\pi d\) (o \(C=2\pi r\)), mentre l'area (A) del cerchio è \(A=\pi r^{2}\)

```python
import math


class Circle:

    def __init__(self,r):
      self.r = r

    def diameter(self):
      """This function returns the diameter of a circle."""
      d = self.r * math.pi
      return d

    def circumference(self):
      """This function returns the circumference of a circle."""
      c = self.r * math.pi * 2
      return c

    def area(self):
      """This function returns the area of a circle."""
      a = self.r ** 2 * math.pi
      return a


my_circle = Circle(4)
print(f"Diametro = ", round(my_circle.diameter(),2))
print(f"Circonferenza = ",round(my_circle.circumference(),2))
print(f"Area = ",round(my_circle.area(),2))
```
