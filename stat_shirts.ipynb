```python
import csv

with open('shirts (1).csv') as csv_file:
    csv_reader = csv.DictReader(csv_file)
    #print(csv_reader.fieldnames)
    print("%s\t%s\t%s\t"%tuple(csv_reader.fieldnames[1:]))

    shirts_stats = {"tot_products":0,
                "val_shirts":0,
                "avg_products":0,
                "prod_byColor":{},
                "prod_bySize":{}
                }

    line_count = 0
    sum_price = 0
    size = set()
    color = set()
    for row in csv_reader:
        print("%s\t%s\t%s\t"%(row["taglia"], row["colore"],row["prezzo"]))
        line_count += 1
        sum_price  += float(row["prezzo"])
        size.add(row["taglia"])
        color.add(row["colore"])

        if row["taglia"] not in shirts_stats["prod_bySize"]:
           shirts_stats["prod_bySize"][row["taglia"]] = 1
        else:
          shirts_stats["prod_bySize"][row["taglia"]] +=1

        if row["colore"] not in shirts_stats["prod_byColor"]:
           shirts_stats["prod_byColor"][row["colore"]] = 1
        else:
          shirts_stats["prod_byColor"][row["colore"]] +=1

        shirts_stats["tot_products"] = line_count
        shirts_stats["val_shirts"] = round(sum_price,3)
        shirts_stats["avg_products"] = round(round(sum_price,3)/line_count,2)
```

```python
print(shirts_stats)
```

## **1. Filtra i proverbi**
Leggi il contenuto del file proverbi.txt e scrivi su di un nuovo file chiamato proverbi_filtrati.txt solo i proverbi che iniziano per vocale o che solo più brevi di 25 caratteri.

```python
proverbs_file.seek(0)
proverbs = proverbs_file.read()
print(proverbs)
```

```python
proverbs_filtred = open("proverbi_filtrati.txt", 'w+')
proverbs_filtred.read()
```

```python
proverbs_file.seek(0)
lista_vocali = ['a','e','i','o','u']
proverbs_filtred_content = []

for proverb in proverbs_file.readlines():
  print("Proverbio: %s" % proverb[:-1]) # rimuoviamo l'ultimo carattere che rappresenta una nuova linea (\n)
  if proverb[0].lower() in lista_vocali or len(proverb)  <= 25:
     proverbs_filtred_content.append(proverb)

with open("proverbi_filtrati.txt", 'w') as f:
    for p in proverbs_filtred_content:
        print(p)
        f.write(p)
```
